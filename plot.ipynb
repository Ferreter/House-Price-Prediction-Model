```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'dataset/cleaned_data.csv'

# Load the dataset
data = pd.read_csv(file_path)

# Pairplot for pairwise relationships between numerical variables
sns.pairplot(data, vars=['price', 'bedrooms', 'bathrooms', 'sqft_living','price_per_sqft','condition'])
plt.show()

```

```python
# Boxplot for price distribution across different numbers of bedrooms
plt.figure(figsize=(10, 6))
sns.boxplot(x='bedrooms', y='price', data=data)
plt.title('Price Distribution Across Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.show()

# Boxplot for price distribution across different numbers of bathrooms
plt.figure(figsize=(10, 6))
sns.boxplot(x='bathrooms', y='price', data=data)
plt.title('Price Distribution Across Bathrooms')
plt.xlabel('Bathrooms')
plt.ylabel('Price')
plt.show()
```

```python

# Scatter plot of price vs. square footage of living area
plt.figure(figsize=(10, 6))
sns.scatterplot(x='sqft_living', y='price', data=data)
plt.title('Price vs. Square Footage of Living Area')
plt.xlabel('Square Footage of Living Area')
plt.ylabel('Price')
plt.show()

# Scatter plot of price vs. condition
plt.figure(figsize=(10, 6))
sns.scatterplot(x='condition', y='price', data=data)
plt.title('Price vs. Condition')
plt.xlabel('Condition')
plt.ylabel('Price')
plt.show()

# Scatter plot of price vs. bedrooms
plt.figure(figsize=(10, 6))
sns.scatterplot(x='bedrooms', y='price', data=data)
plt.title('Price vs. Bedrooms')
plt.xlabel('Bedrooms')
plt.ylabel('Price')
plt.show()


```

```python
# Histogram of price distribution
plt.figure(figsize=(10, 6))
sns.histplot(data['price'], kde=True)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

```
